```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df=pd.read_csv('Titanic_Data.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
#checking if there any missing values
df.isnull().values.any()
```

```python
#summarizing missing values
df.isnull().sum()
```

```python
#total number of missing values
df.isnull().sum().sum()
```

```python
#removing missing values 
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
#1. plotting Age dist using histogram
fig = px.histogram(df, x="Age",title="Age distribution using Histogram")
fig.show()
```

```python
#2. plotting Age dist based on gender using histogram
fig = px.histogram(df, x="Age",color='Sex',title="Age distribution based on gender using Histogram")
fig.show()
```

```python
#3. identifing the correlation within the data
corr=df.corr()
corr
```

```python
#using matrix plot to represent data
fig = px.imshow(df.corr())
fig.show()
```

```python
#4.pivot table of Pclass as the columns and Fare as the values
df.pivot(columns='Pclass',values='Fare')
```

```python
# boxplot to show the distribution of the values in its 3 columns
fig = px.box(df, x="Pclass", y="Fare")
fig.show()
```

```python
#5 Graph the value counts of the number of passengers who survived and did not survive based on gender using a stacked bar graph.
df1=df.value_counts(['Survived','Sex'])
df1
```

```python
Survived=['0','1']
fig = go.Figure(data=[
    go.Bar(name='female', x=Survived, y=[233,81]),
    go.Bar(name='male', x=Survived, y=[468,109])
])
fig.update_layout(barmode='stack')
fig.update_xaxes(title_text='Survived')
fig.update_yaxes(title_text='Value_counts')
fig.update_layout(title_text = "Value counts of the number of passengers who survived and did not survive based on gender")
fig.show()
```

```python
#6 scatter plot of Ages to Fare paid by each passenger based on their gender
fig = px.scatter(df, x="Age", y="Fare", color="Sex",title='scatter plot of Ages to Fare paid by each passenger based on their gender')
fig.show()
```

```python
#7 bubble plot of the ages to the fare paid by each passenger categorizing whether they survived or not
fig = px.scatter(df, x="Age", y="Fare",
                 size="Pclass", color="Survived",
                 hover_name="Name", size_max=40, title=" bubble plot of the ages to the fare paid by each passenger categorizing whether they survived or not")
fig.show()
```
